```python
from bs4 import BeautifulSoup
import requests
```

```python
source_code = requests.get("http://www.statiz.co.kr/stat.php?mid=stat&re=0&ys=2018&ye=2018&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR_ALL_ADJ&o2=TPA&de=1&lr=0&tr=&cv=&ml=2&pa=0&si=&cn=&sn=100").text
soup = BeautifulSoup(source_code)     
```

```python
dat = soup.find_all('span')
```

```python
k = 45           # WAR = 대체 선수 대비 승리 기여
list_war = []  # 타자 WAR 100명 crawling

while  k < 3215:
    data = str(dat[k])
    war = data[50:54]
    list_war.append(war)
    k = k + 32
    
for i in range(len(list_war)): #리스트 내에 문자형을 float형으로 변환
    list_war[i] = float(list_war[i])  
    
list_war
```

```python
print(list_war)
```

```python
k = 72                                  #WPA = 추가한 승리 확률
list_wpa = []                         #WPA 타자 100명 crawling

while  k < 3215:
    data = str(dat[k])
    wpa = data[50:54]
    list_wpa.append(wpa)
    k = k + 32
    
for i in range(len(list_wpa)):   #list내의 문자형을 float형으로 변환
    list_wpa[i] = float(list_wpa[i])
```

```python
list_wpa
```

```python
k = 69                         # Weighted on Base Average(출루율 스케일)
list_woba = []             #타자 wOBA 100명 crawling

while  k < 3215:
    data = str(dat[k])
    woba = data[50:54]
    list_woba.append(woba)
    k = k + 32
    
for i in range(len(list_woba)):   #list 내부의 문자형을 float형으로 변환
    list_woba[i] = float(list_woba[i])
```

```python
list_woba
```
