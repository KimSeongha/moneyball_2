```python
import tensorflow as tf
import pandas as pd
from sklearn.svm import LinearSVC
```

```python
Money =pd.read_csv('../data/kboall2.csv')
data1=Money['WAR']
data2=Money["WPA"]
label=Money["Salary"]
W1 = tf.Variable(tf.random_normal([1]), name = 'weight')
W2 = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf. Variable(tf.random_normal([1]), name='bias')
X1=tf.placeholder(tf.float32, shape=[None])
X2=tf.placeholder(tf.float32, shape=[None])
Y=tf.placeholder(tf.float32, shape=[None])
```

```python
hypothesis = X1 * W1+X2 * W2 + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
```

```python
train=optimizer.minimize(cost)
```

```python
sess=tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, W1_val, W2_val, b_val, _= sess.run([cost, W1, W2, b, train],
                                       feed_dict={X1:data1,X2: data2,Y:label})
    if step % 20 == 0:
        print(step,cost_val,W1_val,W2_val, b_val)
```
